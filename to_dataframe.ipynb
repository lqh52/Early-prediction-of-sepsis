```python
import pandas as pd
import numpy as np
import glob
```

```python
!unzip '/content/drive/MyDrive/Nhiêm trung huyết PROJ /training_setA.zip' -d '/content/drive/MyDrive/Sepsis_predict/'
!unzip '/content/drive/MyDrive/Nhiêm trung huyết PROJ /training_setB.zip' -d '/content/drive/MyDrive/Sepsis_predict/'
```

```python
train_A = glob.glob('/content/drive/MyDrive/Sepsis_predict/training/*.psv')
train_B = glob.glob('/content/drive/MyDrive/Sepsis_predict/training_setB/*.psv')
train = np.concatenate((train_A, train_B))
```

```python
df_list = []
for ind, f in enumerate(train):
    patient_id = f.split('/')[6].split('.')[0].split('p')[-1]
    df = pd.read_csv(f, sep='|')
    df = df.assign(patient=patient_id)

    #print a status update
    if ind % 1000 == 0:
        print(ind)
    
    #append the current parsed file to the list 
    df_list.append(df)


#save all the loaded files into a pickle file
df = pd.concat(df_list)
df = df.reset_index(drop=True)
df.to_pickle('combined.pkl')
```
